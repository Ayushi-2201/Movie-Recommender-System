### Data Preprocessing

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'].replace(" ", ""))
        
    return L
```

```python
movies_dataset['genres'] = movies_dataset['genres'].apply(convert)
```

```python
movies_dataset['keywords'][1]
```

```python
movies_dataset['keywords'] = movies_dataset['keywords'].apply(convert)
```

```python
movies_dataset['cast'][1]
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['order'] < 3:
            L.append(i['name'].replace(" ", ""))
        
    return L
```

```python
movies_dataset['cast'] = movies_dataset['cast'].apply(convert)
```

```python
movies_dataset['crew'][0]
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'].replace(" ", ""))

        
    return L
```

```python
movies_dataset['crew'] = movies_dataset['crew'].apply(convert)
```

```python
movies_dataset.head()
```

```python
movies_dataset['overview'] = movies_dataset['overview'].apply(lambda x:x.split())
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
port_stemmer = nltk.PorterStemmer()
```

```python
def stemming(content):
    # Converting every character to lowercase
    content = [text.lower() for text in content]  # List comprehension to lowercase

    # Removing stopwords
    content = [word for word in content if word not in stopwords.words('english')]

    # Applying stemming
    content = [port_stemmer.stem(word) for word in content]

    # Joining the list to convert it into a string
    content = ' '.join(content)
    
    return content
```

```python
movies_dataset['content'] = movies_dataset['overview'] + movies_dataset['genres'] + movies_dataset['keywords']  + movies_dataset['cast'] + movies_dataset['crew']
```

```python
movies_dataset['content'] = movies_dataset['content'].apply(stemming)
```

```python
movies_dataset.head()
```

```python
movies_dataset.rename(columns={'content': 'tags'}, inplace=True)
```

```python
movies_dataset.head()
```

```python
dataset = movies_dataset[['id', 'title', 'tags']]
```

```python
dataset.head()
```

```python
from collections import Counter
```

```python
word_pool = []
def corpus(text):
    for sentence in text:  
        words = sentence.split() 
        for word in words:
            word_pool.append(word) 
        
corpus(dataset['tags'])
```

```python
frequency = Counter(word_pool)
```

```python
most_common_words = frequency.most_common(50000)
word_pool = [common_word[0] for common_word in most_common_words]
```

```python
word_pool
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(vocabulary= word_pool, max_features=500000)
```

```python
vectors = cv.fit_transform(dataset['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(vectors)
```

```python
if dataset[dataset['title'] == "movie"].empty:
    print("Movie not found")
else:
    movie_index = dataset[dataset['title'] == "movie"].index[0]
    similar_movies = list(enumerate(cosine_sim[movie_index]))
    similar_movies = sorted(similar_movies, key=lambda x:x[1], reverse=True)
    similar_movies = similar_movies[1:6]
    print(similar_movies)
```

```python
def recommend(movie):
    if dataset[dataset['title'] == movie].empty:
        print("Movie not found")
        return
    else:
        index = dataset[dataset['title'] == movie].index[0]
        distances = sorted(list(enumerate(cosine_sim[index])), reverse=True, key=lambda x: x[1])
        for i in distances[1:6]:
            print(dataset.iloc[i[0]].title)
```

```python
recommend("Avatarghdgh")
```
